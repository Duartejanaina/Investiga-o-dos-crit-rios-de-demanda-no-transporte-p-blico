# ANÁLISE E TESTE QUI-QUADRADO  - FREQUÊNCIA NO HEXÁGONO X AUMENTO DE PASSAGEIROS COM RELAÇÃO AO MÊS ANTERIOR

A variável frequência é quantitativa, o primeiro passo é categrizá-la.

```python
# Pacotes
import numpy as np
from scipy.stats import chi2_contingency
import pandas as pd
```

```python
teste_qui_caminho = r"C:\Users\Administrador\Downloads\TCC-20240110T120732Z-001\Dados_UP Gloria\Dados_gerais2.xlsx"
teste_qui = pd.read_excel(teste_qui_caminho)
teste_qui.head()
```

```python
# Identificar os clusters únicos
clusters_unicos = teste_qui['Cluster'].unique()

# Criar um dicionário para armazenar dataframes para cada cluster
dataframes_por_cluster = {cluster: teste_qui[teste_qui['Cluster'] == cluster] for cluster in clusters_unicos}

# Verificar a criação dos dataframes, mostrando a quantidade de linhas de cada um
quantidade_por_cluster = {cluster: len(df) for cluster, df in dataframes_por_cluster.items()}
quantidade_por_cluster
```

```python
# Filtrar o dataframe para o cluster 2
cluster_0_df = teste_qui[teste_qui['Cluster'] == 0]

# Mostrar as primeiras linhas do dataframe filtrado para o cluster 2
cluster_0_df.head()
```

```python
# Filtrar o dataframe para o cluster 2
cluster_1_df = teste_qui[teste_qui['Cluster'] == 1]

# Mostrar as primeiras linhas do dataframe filtrado para o cluster 2
cluster_1_df.head()
```

```python
# Filtrar o dataframe para o cluster 2
cluster_2_df = teste_qui[teste_qui['Cluster'] == 2]

# Mostrar as primeiras linhas do dataframe filtrado para o cluster 2
cluster_2_df.head()
```

```python
# Filtrar o dataframe para o cluster 
cluster_3_df = teste_qui[teste_qui['Cluster'] == 3]

# Mostrar as primeiras linhas do dataframe filtrado para o cluster 
cluster_3_df.head()
```

```python
# Para criar categorias para a variável 'Frequencia_HEX', vamos dividir os dados em quartis dentro do cluster 2.

# Primeiro, vamos encontrar os quartis dentro do cluster 2
quartis_cluster_0 = cluster_0_df['Frequencia_HEX'].quantile([0.25, 0.5, 0.75])
quartis_cluster_0

# Agora vamos usar pd.cut para criar as categorias categóricas baseadas nesses quartis
# Vamos definir as categorias como 'Baixa', 'Média', 'Alta', 'Muito Alta'
categorias = ["Baixa", "Média", "Alta", "Muito Alta"]
cluster_0_df['Frequencia_HEX_Categoria'] = pd.cut(cluster_0_df['Frequencia_HEX'], 
                                                   bins=[-float("inf"), quartis_cluster_0[0.25], quartis_cluster_0[0.5], 
                                                         quartis_cluster_0[0.75], float("inf")], 
                                                   labels=categorias)

# Exibir as primeiras linhas para verificar a categorização
cluster_0_df.head()
```

```python
# Vamos realizar o teste do qui-quadrado entre 'Aumentou' e 'Frequencia_HEX_Categoria'

# Criar uma tabela de contingência
contingency_table_cluster_0 = pd.crosstab(cluster_0_df['Aumentou'], cluster_0_df['Frequencia_HEX_Categoria'])

# Realizar o teste qui-quadrado
chi2_result, p_value, dof, expected_freq = chi2_contingency(contingency_table_cluster_0)

chi2_result, p_value, dof, expected_freq, contingency_table_cluster_0
```
